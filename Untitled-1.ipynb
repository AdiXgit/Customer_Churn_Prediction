```python
import pandas as pd
df = pd.read_csv('Telco-Customer-Churn.csv')
df.head()
df.info()
```

#SMOTE
wre are using SMOTE as it generates synthetic samples for the minority class using k-Nearest Neighbors (k-NN). Instead of copying existing data, it creates new points along the line between existing samples.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler,LabelEncoder

df.drop(columns  = ['customerID'], inplace=True, errors='ignore')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)

categorical_cols = df.select_dtypes(include='object').columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

X = df.drop(columns=['Churn_Yes'])
y = df['Churn_Yes']

smote = SMOTE(sampling_strategy=1.0, random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_smote)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Split data BEFORE applying SMOTE (Stratify ensures class balance)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE only on the training set
smote = SMOTE(sampling_strategy=1.0, random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Standardize the features (fit only on training set)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)  # Transform test set with same scaler

# Function to train and evaluate the model
def train_eval(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\nModel: {model.__class__.__name__}")
    print('Accuracy:', accuracy_score(y_test, y_pred))
    print('Classification Report:\n', classification_report(y_test, y_pred))
    print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))

# Train and evaluate Random Forest
rf_model = RandomForestClassifier(random_state=42, n_estimators=1000)
train_eval(rf_model, X_train_scaled, X_test_scaled, y_train_smote, y_test)

# Train and evaluate XGBoost
xgb_model = XGBClassifier(random_state=42, learning_rate=0.1, n_estimators=1000)
train_eval(xgb_model, X_train_scaled, X_test_scaled, y_train_smote, y_test)
```

```python
#implementing hypeparameter tuning for random forest
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

#first we have to define the hyperparameter grid

rf_param_grid = {
    "n_estimators": [100, 200, 300, 500],
    "max_depth": [10, 20, 30, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "bootstrap": [True, False]
}

#initialize the random forest classifier
rf_model = RandomForestClassifier(random_state=42,class_weight = 'balanced')
rf_random = RandomizedSearchCV(rf_model, param_distributions=rf_param_grid, 
                               n_iter=20, cv=5, scoring="f1", random_state=42, n_jobs=-1, verbose=2)

rf_random.fit(X_train_scaled, y_train_smote)
print(rf_random.best_params_)
best_rf = rf_random.best_estimator_


xgb_param_grid = {
    "n_estimators": [100, 200, 300, 500],
    "learning_rate": [0.01, 0.05, 0.1, 0.2],
    "max_depth": [3, 5, 7, 10],
    "subsample": [0.6, 0.8, 1.0],
    "colsample_bytree": [0.6, 0.8, 1.0],
    "gamma": [0, 0.1, 0.2, 0.3]
}

xgb_model = XGBClassifier(random_state=42, scale_pos_weight=len(y_train[y_train == 0]) / len(y_train[y_train == 1]))

xgb_random = RandomizedSearchCV(xgb_model, param_distributions=xgb_param_grid, 
                                n_iter=20, cv=5, scoring="f1", random_state=42, n_jobs=-1, verbose=2)

xgb_random.fit(X_train, y_train)

print("Best Hyperparameters for XGBoost:", xgb_random.best_params_)

# Train the best model
best_xgb = xgb_random.best_estimator_
```

```python
best_rf = rf_random.best_estimator_
best_rf.fit(X_train_scaled, y_train_smote)


```

```python
best_xgb = xgb_random.best_estimator_
best_xgb.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions
y_pred_rf = best_rf.predict(X_test_scaled)  # Use scaled test data for RF
y_pred_xgb = best_xgb.predict(X_test)  # Use raw test data for XGB

# Evaluate Random Forest
print("Random Forest Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))

# Evaluate XGBoost
print("\nXGBoost Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
```
